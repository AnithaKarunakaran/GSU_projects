```python
#getting columns in same order as required to rename the columns
data.columns

data = data[['Ticker', 'Company', 'Title', 'Date', 'Time', 'Section', 'Speech','Participant name', 'Company_Participants_type', 'Participant_Organization']]
data.columns
```

```python
#change column names
data.columns= ['Ticker', 'Company', 'Title', 'Conference_Date', 'Conference_Time', 'Section', 'Speech','Participant_name','Participant_type', 'Participant_Organization']
```

```python
data.columns
```

```python
data.head(2)
```

```python
### Assign a number to the group
df1 = data.copy()
 
#Conference_id starting from 100
df1['Conference_Id'] = df1.groupby(['Ticker','Company','Title','Conference_Date','Conference_Time']).ngroup()+100

#Participant_id starting from 1000
#df1['Participant_id'] = df1.groupby(['Ticker','Company','Title','Conference_Date','Conference_Time','Participant_name','Participant_type', 'Participant_Organization']).ngroup()+10
df1['Participant_id'] = df1.groupby(['Ticker','Company','Title','Conference_Date','Conference_Time','Participant_name']).ngroup()+1000

#this keeps the order of the speech as is
df1['Speech_id'] = np.arange(len(df1)) + 1

df1.head(5)
```

```python
df1.columns
```

##### Creating Conference table

Columns are:
+ Conference_id
+ Title
+ Conference_Date
+ Conference_Time
+ Ticker
+ Company
+ Quarter

```python
Conference_df = pd.DataFrame()  
Conference_df = df1[["Conference_Id","Title","Conference_Date","Conference_Time","Ticker","Company"]].drop_duplicates()

# convert the 'Date' column to datetime format
Conference_df['Conference_Date'] = pd.to_datetime(Conference_df['Conference_Date'])

#create quarter column from date
Conference_df['Conference_Quarter'] = Conference_df['Conference_Date'].dt.quarter
#adding Q in-front of quarter column
Conference_df['Conference_Quarter'] = 'Q' + Conference_df['Conference_Quarter'].astype(str)

print(Conference_df.shape)

Conference_df.head(20)
```

#### Creating Participant table

Columns are:
+ Participant_id
+ Participant name
+ Participant type
+ Participant Organization

```python
Participant_df = pd.DataFrame()  
Participant_df = df1[['Participant_id','Participant_name','Participant_type', 'Participant_Organization']].drop_duplicates()

print(Participant_df.shape)

Participant_df.head(20)
```

#### Creating Conference_Participant table

Columns are:
+ Conference_Id
+ Participant_id

```python
Conference_Participant_df = pd.DataFrame()  
Conference_Participant_df = df1[['Conference_Id','Participant_id']].drop_duplicates()

print(Conference_Participant_df.shape)
Conference_Participant_df.head(20)
```

#### Creating Speech table

Columns are:
+ Speech_id
+ Conference_Id
+ Participant_id
+ Section
+ Speech

```python
Speech_df = pd.DataFrame()  
Speech_df = df1[['Speech_id','Conference_Id', 'Participant_id','Section', 'Speech']].drop_duplicates()

print(Speech_df.shape)
Speech_df.tail(20)
```

```python
#remove punctuations from Speech
import re

Speech_df['Speech'] = Speech_df['Speech'].str.replace('[^\w\s]','')
```

```python
Speech_df.tail(20)
```

```python
Conference_df.to_csv("Conference.csv",index=False)

Participant_df.to_csv("Participant.csv",index=False)

Conference_Participant_df.to_csv("Conference_Participant.csv",index=False)

Speech_df.to_csv("Speech.csv",index=False)
```
